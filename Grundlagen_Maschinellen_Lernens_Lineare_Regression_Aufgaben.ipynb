```python
import plotly.graph_objects as go
import numpy as np

# Annahme: df ist bereits definiert und enthält die Daten
x = df['Teen']

# Konvertiere 'sex' zu Kategorie und erhalte Codes
sex_cat = df['State'].astype('category')
y = sex_cat.cat.codes

# Konvertiere 'day' zu Kategorie und erhalte Codes
z=x = df['Mort']

# Erstelle ein Mapping von Codes zu Kategorien für die y-Achse
y_codes = np.unique(y)
y_categories = sex_cat.cat.categories.tolist()
y_tickvals = y_codes
y_ticktext = [y_categories[code] for code in y_codes]

# Erstelle ein Mapping von Codes zu Kategorien für die z-Achse


# Erstelle einen 3D-Scatter-Plot
fig = go.Figure(data=go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        #color=density,  # Optional: Farbe basierend auf Dichte oder einer anderen Variablen
        colorscale='Viridis',
        opacity=0.8
    )
))

# Aktualisiere das Layout des Plots
fig.update_layout(
    template='plotly_white',
    title='3D Scatter Plot of Tips by Gender and Day',
    width=1000,
    height=800,
    scene=dict(
        xaxis_title='Teen',
        yaxis_title='State',
        zaxis_title='Mort',
        yaxis=dict(
            tickmode='array',
            tickvals=y_tickvals,
            ticktext=y_ticktext,
            range=[min(y)-0.5, max(y)+0.5]
        ),

    )
)

fig.show()
```
